```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.stripplot(y='tip',data=tips,jitter=True) # Jitter is used spread the data points as shown below
plt.ylabel('Tips($)')                        # By default Jitter will be True
plt.show()                         # For 1 dimension or using one column
```

```python
sns.stripplot(x='day',y='tip',data=tips) #By using 2 dimesion o2 2 column
```

```python
sns.swarmplot(x='day',y='tip',data=tips)
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex') #Hue is used to differentiate based on that particular column
```

```python
sns.boxplot(x='day',y='tip',data=tips)
```

```python
sns.violinplot(x='day',y='tip',data=tips)
```

```python
#To plot more than 1 plot

plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.subplot(1,2,2)
sns.stripplot(x='day',y='tip',data=tips)
```

```python
#To create a joint plot where 2 will be present in one plot itself
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#To see the plots for all the groups
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```
